```python
def numero_binario(numero):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    if (type(numero)!=int):
        return None
    if (numero < 0):
        return None
    base = 2
    lista_base = [1]
    lista_activos = [0]
    while(base <= numero):
        lista_base.append(base)
        lista_activos.append(0)
        base = base * 2
    i = len(lista_base) - 1
    numero_binario = 0
    while(i >= 0):
        if (numero_binario + lista_base[i] <= numero):
            numero_binario += lista_base[i]
            lista_activos[i]=1
        i-=1

    i = len(lista_base) - 1
    numero_binario_final = ''
    while(i >= 0):
        numero_binario_final = numero_binario_final + str(lista_activos[i])
        i-=1
    return int(numero_binario_final)
```

```python
# Este código es una implementación de una función llamada numero_binario que toma un número entero como parámetro y devuelve su representación en binario. 
# Veamos paso a paso cómo funciona:

# Esta parte define la función numero_binario que toma un parámetro numero. 
# La documentación en triple comillas explica brevemente lo que hace la función y los requisitos de entrada y salida.

def numero_binario(numero):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    
# Aquí se realizan las primeras comprobaciones:
# Se verifica si el tipo del parámetro numero no es un entero. Si no lo es, devuelve None.
# Se verifica si el número es menor que 0. Si lo es, también devuelve None.
    
    if (type(numero)!=int):
        return None
    if (numero < 0):
        return None
    
# Se inicializan algunas variables importantes:
# base se establece en 2, ya que estamos convirtiendo a binario.
# lista_base se inicia con [1] ya que la primera potencia de 2 es 2^0.
# lista_activos se inicia con [0], que luego se actualizará para indicar qué potencias de 2 están presentes en la representación binaria del número.
    
    base = 2
    lista_base = [1]
    lista_activos = [0]

# Este bucle calcula potencias de 2 crecientes (hasta que superen el número a convertir) y las agrega tanto a lista_base como a lista_activos.

    while(base <= numero):
        lista_base.append(base)
        lista_activos.append(0)
        base = base * 2
        
#   Aquí se calcula el número binario utilizando las potencias de 2 generadas anteriormente. 
# Se recorre lista_base de derecha a izquierda, sumando potencias de 2 al número binario si es posible sin exceder el número original. 
# También se actualiza lista_activos para indicar qué potencias de 2 se incluyen en la representación binaria.  
        
    i = len(lista_base) - 1
    numero_binario = 0
    while(i >= 0):
        if (numero_binario + lista_base[i] <= numero):
            numero_binario += lista_base[i]
            lista_activos[i]=1
        i-=1
        
#  Este bucle construye una cadena de dígitos binarios utilizando los valores de lista_activos, empezando desde la potencia de 2 más alta.

    i = len(lista_base) - 1
    numero_binario_final = ''
    while(i >= 0):
        numero_binario_final = numero_binario_final + str(lista_activos[i])
        i-=1
        
# Finalmente, se devuelve el número binario representado como un entero después de convertir la cadena binaria final a un entero.
        
    return int(numero_binario_final)
```

```python
# Esta línea define una función llamada numero_binario_fraccionario que toma un parámetro llamado numero_decimal, 
# que se espera que sea un número de punto flotante que represente una fracción en decimal.

def numero_binario_fraccionario(numero_decimal):
    
    # Aquí se realiza una verificación para asegurarse de que el parámetro numero_decimal sea realmente un número de punto flotante. 
    # Si no lo es, la función devuelve None.
    
    if (type(numero_decimal) != float):
        return None
    
    # Este bloque verifica si numero_decimal está dentro del rango válido para una fracción, es decir, entre 0 y 1. 
    # Si numero_decimal está fuera de este rango, la función devuelve None.
    
    if ((numero_decimal > 1) or (numero_decimal < 0)):
        return None
    
    # Si numero_decimal es exactamente igual a 0, la función devuelve simplemente 0.    
    
    elif (numero_decimal == 0):
        return 0
    
    # Aquí se inicializan algunas variables importantes: lim_periodico es el límite para evitar bucles infinitos en caso de fracciones periódicas, 
    # parte_entera se inicializa en 0 (ya que estamos tratando con fracciones) y lista_binario es una lista vacía que almacenará los dígitos binarios de la fracción. 
    # i se inicializa en 0 y se utilizará para llevar la cuenta de las iteraciones.
    
    lim_periodico = 24
    parte_entera = 0
    lista_binario = []
    i = 0
    
    # Este bucle while se ejecuta mientras numero_decimal sea mayor que 0 (es decir, mientras quede una parte fraccionaria por convertir) 
    # y i sea menor que lim_periodico (para evitar bucles infinitos). 
    # En cada iteración, se agrega el dígito binario correspondiente al número actual multiplicado por 2 a la lista lista_binario. 
    # Luego, se actualiza numero_decimal restando la parte entera del número (obtenida mediante int(numero_decimal * 2)) y se incrementa i.
    
    while((numero_decimal > 0) and (i < lim_periodico)):
        lista_binario.append(int(numero_decimal * 2))
        numero_decimal = numero_decimal * 2 - (int(numero_decimal * 2))
        i+=1
        
    # Después del bucle, se inicializa la cadena numero_binario con el valor inicial '0.' que indicará la parte fraccionaria del número binario. 
    # Luego, se itera sobre cada elemento en lista_binario y se concatenan los dígitos binarios a numero_binario.
        
    numero_binario = '0.'
    for e in lista_binario:
        numero_binario = numero_binario + str(e)
        
    # Finalmente, se devuelve la representación binaria de la fracción como una cadena de caracteres.

    return(numero_binario)    
```

```python
# Homework Carla (explicacion)

'''definimos la funcion que toma 2 parametros:numero, que es el numero decimal que queremos convertir 
y presicion, que es la cantidad de bits despues del punto decimal en la representacion binaria. Por defecto presicion se establece en 8'''
def decimal_a_binario_fraccion(numero, presicion=8):

    '''extraemos la parte entera del numero utilizando la funcion int 
    y almacenamos la parte decimal restando la parte entera del numero original'''
    parte_entera = int(numero)
    parte_decimal = numero - parte_entera

    '''convertimos la parte entera a su representacion binaria utilizando la funcion bin(). 
    La parte [2:] es para eliminar los primeros dos caracteres que indican que es una representacion binaria'''
    # la funcion bin() en pythontoma un numero entero como argumento y devuelve una cadena que representa su equivalente binario.
    # ej: bin(5)devolvera '0b101', donde 0b indica que es una representacion binaria de pythony 101 es el equivalente binario de 5
    # [2:] se utiliza para eliminar los primeros dos caracteres 0b de la cadena generada por bin para obtener solo la representacion binaria del numero

    parte_entera_binaria = bin(parte_entera)[2:]

    '''iniciamos una cadena vacia para almacenar la representacion binaria de la parte decimal.
    luego iteramos presicion para calcular los bits de la parte decimal .
    en cada iteracion multiplicamos la parte decimal por 2, tomamos la parte entera del resultado como el siguiente bit,
     lo convertimos a cadena y loagregamos a parte_decimal_binaria.
    despues restamos la parte entera del resultado de la parte decimal para obtener la parte decimal para la siguiente iteracion  '''

    parte_decimal_binaria = ""
    for _ in range(presicion):
        parte_decimal *= 2
        bit = int(parte_decimal)
        parte_decimal_binaria += str(bit)
        parte_decimal -= bit

        '''concatenamos la presicion binaria de la parte entera con un punto 
        y la representacion binaria de la parte decimalpara obtener el resultado final y lo devolvemos'''
    
    return parte_entera_binaria + "." + parte_decimal_binaria
```
